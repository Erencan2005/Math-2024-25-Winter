6. **Profit Function:**

The profit function is given as:
\[
P(u) = -2u^2 + 50u - 300
\]

To find the number of units \( u \) that maximize profit, calculate the critical points by finding the derivative and setting it to zero:
\[
P'(u) = -4u + 50
\]
Set \( P'(u) = 0 \):
\[
-4u + 50 = 0 \implies u = 12.5
\]

Verify it is a maximum using the second derivative:
\[
P''(u) = -4
\]
Since \( P''(u) < 0 \), the profit function has a maximum at \( u = 12.5 \).

Thus, the number of units that maximize profit is:
\[
u = 12.5
\]

7. **Largest Rectangle Enclosed by 10 Meters of String:**

Let the length of the rectangle be \( l \) and the width \( w \). The perimeter is given as:
\[
2l + 2w = 10 \implies l + w = 5
\]

The area of the rectangle is:
\[
A = l \cdot w
\]
Substitute \( w = 5 - l \):
\[
A = l(5 - l) = 5l - l^2
\]

Find the critical points by taking the derivative and setting it to zero:
\[
A'(l) = 5 - 2l
\]
Set \( A'(l) = 0 \):
\[
5 - 2l = 0 \implies l = 2.5
\]

Verify it is a maximum using the second derivative:
\[
A''(l) = -2
\]
Since \( A''(l) < 0 \), the area is maximized at \( l = 2.5 \).

Thus, the rectangle's dimensions are:
\[
l = 2.5, \, w = 5 - 2.5 = 2.5
\]

8. **Extremum of \( f(x) = x^2 + 3x - 5 \):**

Find the derivative:
\[
f'(x) = 2x + 3
\]
Set \( f'(x) = 0 \):
\[
2x + 3 = 0 \implies x = -\frac{3}{2}
\]

Verify the nature of the critical point using the second derivative:
\[
f''(x) = 2
\]
Since \( f''(x) > 0 \), the function has a minimum at \( x = -\frac{3}{2} \).

The minimum value is:
\[
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5 = \frac{9}{4} - \frac{9}{2} - 5 = -\frac{29}{4}
\]

Thus, the minimum value is:
\[
x = -\frac{3}{2}, \, f(x) = -\frac{29}{4}
\]

9. **Extremum of \( f(x) = \frac{x^2 + 2x + 1}{x - 1} \):**

Rewrite the function:
\[
f(x) = \frac{(x + 1)^2}{x - 1}
\]

Find the derivative using the quotient rule:
\[
f'(x) = \frac{(x - 1) \cdot 2(x + 1) - (x + 1)^2 \cdot 1}{(x - 1)^2}
\]
Simplify:
\[
f'(x) = \frac{2(x + 1)(x - 1) - (x + 1)^2}{(x - 1)^2} = \frac{(x + 1)(2x - 2 - (x + 1))}{(x - 1)^2}
\]
\[
f'(x) = \frac{(x + 1)(x - 3)}{(x - 1)^2}
\]

Set \( f'(x) = 0 \):
\[
(x + 1)(x - 3) = 0 \implies x = -1, \, x = 3
\]

Check the domain: \( x \neq 1 \).

Verify the nature of the critical points using the second derivative or sign changes in \( f'(x) \). Analyze behavior around \( x = -1 \) and \( x = 3 \).

Thus, the critical points are:
\[
x = -1, \, x = 3
