````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbol
x = sp.symbols('x')

# Compute integrals
integrals = {
    "int(1 dx)": sp.integrate(1, x),
    "int(x^2 + 2 dx)": sp.integrate(x**2 + 2, x),
    "int(2sin(x) dx)": sp.integrate(2 * sp.sin(x), x),
    "int(3/x dx)": sp.integrate(3/x, x),
    "int(1/x^2 dx)": sp.integrate(1/x**2, x),
    "int((1/3)x^4 - 5 dx)": sp.integrate((1/3)*x**4 - 5, x),
    "int(sin^2 x + cos^2 x dx)": sp.integrate(sp.sin(x)**2 + sp.cos(x)**2, x),
    "int(5 sin x + 3e^x dx)": sp.integrate(5 * sp.sin(x) + 3 * sp.exp(x), x),
    "int(cube_root(x) dx)": sp.integrate(x**(1/3), x),
    "int(sqrt(10x) dx)": sp.integrate(sp.sqrt(10*x), x),
    "int(cos(5/2 x + 3) dx)": sp.integrate(sp.cos((5/2)*x + 3), x),
    "int(cos(ln(x))/x dx)": sp.integrate(sp.cos(sp.ln(x))/x, x),
    "int(x ln(x) dx)": sp.integrate(x * sp.ln(x), x),
    "int(x e^x dx)": sp.integrate(x * sp.exp(x), x),
}

# Print integrals
for key, value in integrals.items():
    print(f"{key} = {value}")

# Definite integrals over [0, pi]
definite_integrals = {
    "int(2x+1, [0,pi])": sp.integrate(2*x + 1, (x, 0, sp.pi)),
    "int(x^2, [0,pi])": sp.integrate(x**2, (x, 0, sp.pi)),
}

print("\nDefinite Integrals:")
for key, value in definite_integrals.items():
    print(f"{key} = {value}")

# Area bounded by curves x=1, x=2, y=0, y=x^2+1
area_bounded = sp.integrate(x**2 + 1, (x, 1, 2)) - sp.integrate(0, (x, 1, 2))
print(f"\nArea bounded by x=1, x=2, y=0, y=x^2+1: {area_bounded}")

# Area under sine curve over [0, pi]
area_sine = sp.integrate(sp.sin(x), (x, 0, sp.pi))
print(f"\nArea under sine curve over [0, pi]: {area_sine}")

# Plotting definite integrals
x_vals = np.linspace(0, np.pi, 400)
y_f = 2*x_vals + 1
y_g = x_vals**2

def plot_function(x_vals, y_vals, label, color):
    plt.plot(x_vals, y_vals, label=label, color=color)
    plt.fill_between(x_vals, y_vals, alpha=0.3, color=color)

plt.figure(figsize=(10, 6))
plot_function(x_vals, y_f, 'f(x) = 2x + 1', 'blue')
plot_function(x_vals, y_g, 'g(x) = x^2', 'red')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Definite Integrals over [0, pi]")
plt.show()
````
