\subsection*{1. Taylor Series Expansions}

\textbf{(a) Taylor Series for $f(x) = \cos(x)$ at $x = 0$ up to 4th degree:}

The Taylor series expansion of $\cos(x)$ around $x=0$ is given by:
\begin{align*}
    \cos(x) &= \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!} x^{2n} \\
            &\approx 1 - \frac{x^2}{2!} + \frac{x^4}{4!}.
\end{align*}

In Geogebra, define:
```

 taylor_f(x) = 1 - x^2/2 + x^4/24
```

\textbf{(b) Taylor Series for $h(x) = \frac{1}{1-x}$ at $x = 0$ up to 4th degree:}

The Taylor series for $\frac{1}{1-x}$ around $x=0$ is:
\begin{align*}
    \frac{1}{1-x} &= \sum_{n=0}^{\infty} x^n \\
                   &\approx 1 + x + x^2 + x^3 + x^4.
\end{align*}

In Geogebra, define:
```
 h(x) = 1/(1-x)
 taylor_h(x) = 1 + x + x^2 + x^3 + x^4
```

\textbf{(c) Taylor Series for $g(x) = \sin(x)$ at $x = \pi$ up to 4th degree:}

Using the Taylor expansion around $x = \pi$:
\begin{align*}
    g(x) &= \sin(x) \\
    g(\pi + h) &= \sin(\pi + h) = -\sin(h) \\
    &= -\left(h - \frac{h^3}{3!}\right)\\
    &= - (x - \pi) + \frac{(x - \pi)^3}{6}.
\end{align*}

In Geogebra, define:
```

 taylor_g(x) = -(x - π) + (x - π)^3/6
```

\subsection*{2. Tangent Line to $f(x) = e^{\sin(x)}$ at $x_0 = \pi$}

The derivative of $f(x) = e^{\sin(x)}$ is:
\begin{align*}
    f'(x) = e^{\sin(x)} \cos(x).
\end{align*}
Evaluating at $x_0 = \pi$:
\begin{align*}
    f(\pi) &= e^{\sin(\pi)} = e^0 = 1,\\
    f'(\pi) &= e^{\sin(\pi)} \cos(\pi) = 1 \cdot (-1) = -1.
\end{align*}

Thus, the equation of the tangent line is:
\begin{align*}
    y = -1 (x - \pi) + 1 = -x + \pi + 1.
\end{align*}

````python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols
x = sp.symbols('x')

# Functions
f_x = sp.cos(x)
h_x = 1 / (1 - x)
g_x = sp.sin(x)

# Taylor series expansions
f_taylor = sp.series(f_x, x, 0, 5).removeO()
h_taylor = sp.series(h_x, x, 0, 5).removeO()
g_taylor = sp.series(g_x, x, sp.pi, 5).removeO()

print("Taylor series expansions:")
print(f"cos(x) ≈ {f_taylor}")
print(f"1/(1-x) ≈ {h_taylor}")
print(f"sin(x) around pi ≈ {g_taylor}")

# Plot Taylor series approximations
x_vals = np.linspace(-2, 2, 400)
plt.figure(figsize=(12, 6))

# Function to plot
for func, taylor, title in zip([f_x, h_x, g_x], [f_taylor, h_taylor, g_taylor],
                               ["cos(x) at x=0", "1/(1-x) at x=0", "sin(x) at x=pi"]):
    plt.plot(x_vals, [sp.N(func.subs(x, val)) for val in x_vals], label=f"{title} original", linestyle='dashed')
    plt.plot(x_vals, [sp.N(taylor.subs(x, val)) for val in x_vals], label=f"{title} Taylor approx")

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.title("Taylor Series Approximations")
plt.show()

# Tangent Line to f(x) = e^sin(x) at x0 = pi
f_x = sp.exp(sp.sin(x))
x0 = sp.pi
f_prime = sp.diff(f_x, x)
tangent_line = f_prime.subs(x, x0) * (x - x0) + f_x.subs(x, x0)

print(f"Tangent line at x0 = pi: y = {sp.simplify(tangent_line)}")
````
